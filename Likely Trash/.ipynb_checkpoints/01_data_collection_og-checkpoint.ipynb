```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
base_url = 'https://www.praisecharts.com/blog/top-40-worship-songs-of-'
year_list = range(2004, 2021)
metadata_urls = []

for year in year_list:
    songlist_url = base_url + str(year) + '/'
    songlist_res = requests.get(songlist_url)
    songlist_soup = BeautifulSoup(songlist_res.content, 'lxml')
    top_40 = songlist_soup.find('div', {'class': 'blog-entry'})
    for song in top_40.find_all('h4'):
        metadata_urls.append(song.find('a').attrs['href'])
        
meta_data
```

```python
song_dicts = []
for url in tqdm(metadata_urls):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'lxml')
    
    lyrics_sidebar = soup.find('section', {'id': 'sidebar'}).find('div', {'class': 'lyrics'})
    song_lyrics = []
    try:
        for lyric in lyrics_sidebar.find_all('div'):
            if lyric.attrs['data-line'] not in ['Verse ', 'Verse 1', 'Verse 2', 'Verse 3',
                                                'Chorus', 'Chorus ', 'Chorus 1', 'Chorus 2',
                                                'Chorus 3', 'Bridge', 'Bridge ', 'Vamp ',
                                               'tag',]:
                song_lyrics.append(lyric.text.lower())   
    except:
        print(f'No lyrics found at {url}')
        
    metadata_table = soup.find('table')
    metadata = {}
    for i in metadata_table.find_all('tr'):
        #metadata = {}
        metadata[i.find_all('td')[0].text[:-1]] = i.find_all('td')[1].text.rstrip()
    
    metadata['lyrics'] = song_lyrics
    metadata['song'] = soup.find('h1', {'id': 'product-detail-title'}).text.lstrip().rstrip()
    song_dicts.append(metadata)
```

```python
pd.DataFrame(song_dicts).to_csv('song_list')
```
